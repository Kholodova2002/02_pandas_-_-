```python
import pandas as pd
print('ЗАДАНИЕ 1.1')
print('--------------------')
recipes = pd.read_csv('recipes_sample.csv')
reviews = pd.read_csv('reviews_sample.csv', index_col=0)
print('ЗАДАНИЕ 1.2')
print('--------------------')
print("recipes:")
print(recipes.info())

print("reviews:")
print(reviews.info())

print('ЗАДАНИЕ 1.3')
print('--------------------')

print("Доля строк, содержащих пропуски в таблице recipes:")
print(recipes.isnull().mean())

print("Доля строк, содержащих пропуски в таблице reviews:")
print(reviews.isnull().mean())

print('ЗАДАНИЕ 1.4')
print('--------------------')

'''Для рассчета среднего значения для каждого из числовых столбцов в таблице
recipes можно воспользоваться методом describe(), который выводит основные
статистические характеристики таблицы, включая среднее значение для числовых столбцов.
'''

print("основные статистические характеристики в таблице recipes:")
print(recipes.describe())

print("основные статистические характеристики в таблице recipes:")
print(recipes['n_ingredients'].mean())


print('ЗАДАНИЕ 1.5')
print('--------------------')

# выбор 10 случайных названий рецептов
recipe_names = recipes['name'].sample(n=10)

print(recipe_names)

print('ЗАДАНИЕ 1.6')
print('--------------------')

# изменение индекса
reviews = reviews.reset_index(drop=True)

print(reviews.head())

print('ЗАДАНИЕ 1.7')
print('--------------------')

# выбор рецептов с временем выполнения не более 20 минут и не более 5 ингредиентами
selected_recipes = recipes[(recipes["minutes"] <= 20) & (recipes["n_ingredients"] <= 5)]

# вывод информации о выбранных рецептах
print(selected_recipes[["name", "minutes", "n_ingredients"]])

print('ЗАДАНИЕ 2.1')
print('--------------------')

# преобразование столбца submitted в формат времени
recipes["submitted"] = pd.to_datetime(recipes["submitted"])

print(recipes["submitted"])

print('ЗАДАНИЕ 2.2')
print('--------------------')

# выбор рецептов, добавленных не позже 2010 года
selected_recipes = recipes[recipes["submitted"].dt.year <= 2010]

# вывод информации о выбранных рецептах
print(selected_recipes[["name", "submitted"]])

print('ЗАДАНИЕ 3.1')
print('--------------------')


# добавление столбца с длиной описания рецепта
recipes["description_length"] = recipes["description"].str.len()

# вывод первых строк таблицы с новым столбцом
print(recipes.head())

print('ЗАДАНИЕ 3.2')
print('--------------------')

recipes['name'] = recipes['name'].str.title()
print(recipes['name'])

print('ЗАДАНИЕ 3.3')
print('--------------------')

# убираем лишние пробелы между словами
recipes['name'] = recipes['name'].str.replace('\s+', ' ', regex=True)

# разбиваем название на слова и считаем их количество
recipes['name_word_count'] = recipes['name'].str.split().apply(len)

print(recipes['name_word_count'])

print('ЗАДАНИЕ 4.1')
print('--------------------')
# количество пецептов
recipe_counts = recipes['contributor_id'].value_counts()
# максимальное количесвто рецепов
max_contributor = recipe_counts.idxmax()
print(max_contributor)

print('ЗАДАНИЕ 4.2')
print('--------------------')

# расчет среднего рейтинга каждого рецепта
mean_ratings = reviews[reviews['rating'] > 0].groupby('recipe_id')['rating'].mean()

# количество рецептов без отзывов
no_reviews_mask = ~recipes['id'].isin(reviews[reviews['rating'] > 0]['recipe_id'].unique())
no_reviews_count = no_reviews_mask.sum()

print(no_reviews_count)

print('ЗАДАНИЕ 4.3')
print('--------------------')

recipes['year_created'] = pd.to_datetime(recipes['submitted']).dt.year
recipe_count_by_year = recipes['year_created'].value_counts()
print(recipe_count_by_year)

print('ЗАДАНИЕ 5.1')
print('--------------------')

df = pd.merge(reviews[['recipe_id', 'user_id', 'rating']], recipes[['id', 'name']], left_on='recipe_id', right_on='id', how='inner')
df.drop('recipe_id', axis=1, inplace=True)
print(df)

print('ЗАДАНИЕ 5.2')
print('--------------------')

recipe_review_count = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id', how='left')\
                       .groupby(['recipe_id', 'name'])['rating']\
                       .count()\
                       .reset_index(name='review_count')\
                       .fillna(0)
print(recipe_review_count)

print('ЗАДАНИЕ 5.3')
print('--------------------')

merged = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id')
merged['year'] = pd.DatetimeIndex(merged['submitted']).year
grouped = merged.groupby('year')[['rating']].mean()
min_rating_year = grouped['rating'].idxmin()
print("Год с наименьшим средним рейтингом:", min_rating_year)

print('ЗАДАНИЕ 6.1')
print('--------------------')

recipes = recipes.sort_values('name_word_count', ascending=False)

recipes[['description_length', 'name', 'name_word_count']].to_csv('6.1.csv', index=False)

print('ЗАДАНИЕ 6.2')
print('--------------------')

df.to_excel(pd.ExcelWriter('6.2.xlsx'), sheet_name='Рецепты с оценками')
recipe_review_count.to_excel(pd.ExcelWriter('6.2.xlsx'), sheet_name='Количество отзывов по рецептам')

```
